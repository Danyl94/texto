<b>Corpus de Gutenberg</b>
<p>NLTK incluye una pequea seleccin de textos del archivo de texto electrnico del Proyecto Gutenberg, que contiene unos 25 000 libros electrnicos gratuitos, alojados en http://www.gutenberg.org/.</p>

```python
# import nltk
# nltk.corpus.gutenberg.fileids()

from nltk.corpus import gutenberg
gutenberg.fileids()
```

```python
# Escojamos el primero de estos textos, Emma de Jane Austen
emma = gutenberg.words('austen-emma.txt')
len(emma)
```

SInopsis: Emma se enfrenta a un vaco en su vida y con un gran dilema: cmo ayudar a los dems a tener una vida tan perfecta como la suya. En contra del consejo de Knightley, busca posibles novios para su nueva amiga, Harriet Smith, una joven sencilla y modesta, alejada del estilo de vida de la alta sociedad.

Texto original: Emma (1816) en Wikisource

Fecha de publicacin: 23 de diciembre de 1815

Gnero: Novela

```python
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid)) # nmero de caracteres
    num_words = len(gutenberg.words(fileid)) # nmero de palabras
    num_sents = len(gutenberg.sents(fileid)) # nmero de oraciones
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid))) # nmero de vocabulario
    
    # Porcentaje de nmero de caracteres por palabras, palabras por oraciones y palabras por vocabulario
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

<p>La funcin <b>raw()</b> nos da el contenido del archivo sin ningn procesamiento lingstico. Entonces, por ejemplo, <b>len (gutenberg.raw ('blake-poems.txt'))</b> nos dice cuntas letras aparecen en el texto, incluidos los espacios entre palabras.</p>
    
<p>La funcin <b>sents()</b> divide el texto en sus oraciones, donde cada oracin es una lista de palabras</p>

```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content) / len(text)

content_fraction(text1) # text1: Moby Dick by Herman Melville 1851
```
